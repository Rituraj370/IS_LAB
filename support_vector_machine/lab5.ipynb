```python
import pandas as pd
import numpy as np
```

```python
db = pd.read_csv('Social_Network_Ads.xls')
db.head()
```

```python
db.drop('User ID', inplace=True, axis=1)
db.head()
```

```python
Gen = pd.get_dummies(db.Gender)
Gen
```

```python
db = pd.concat([db, Gen], axis="columns")
db
```

```python
db.drop('Gender', inplace=True, axis=1)
db
```

```python
Pur_col = db.pop('Purchased')
```

```python
db.insert(4, 'Purchased', Pur_col)
db
```

```python
X = db.iloc[:, 0:4]
Y = db.iloc[:, 4]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=8)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Predicted y values are: ",*y_pred)
```

```python
from sklearn import metrics
acc = metrics.accuracy_score(y_test, y_pred)
print("Accuracy of our model is: ",acc*100)
```

```python
print("Precision:", metrics.precision_score(y_test, y_pred))
print("Recall:", metrics.recall_score(y_test, y_pred))
```
